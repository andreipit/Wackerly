```
##### Problem 1.13
plt.clf()
%clear
output.clear()

X = np.array([8.9, 12.4, 8.6, 11.3, 9.2, 8.8, 35.1, 6.2, 7.0,
7.1, 11.8, 10.7, 7.6, 9.1, 9.2, 8.2, 9.0, 8.7,
9.1, 10.9, 10.3, 9.6, 7.8, 11.5, 9.3, 7.9, 8.8,
8.8, 12.7, 8.4, 7.8, 5.7, 10.5, 10.5, 9.6, 8.9,
10.2, 10.3, 7.7, 10.6, 8.3, 8.8, 9.5, 8.8, 9.4])

# Removed the 35.1 outlier.
#X = np.array([8.9, 12.4, 8.6, 11.3, 9.2, 8.8, 6.2, 7.0,
#7.1, 11.8, 10.7, 7.6, 9.1, 9.2, 8.2, 9.0, 8.7,
#9.1, 10.9, 10.3, 9.6, 7.8, 11.5, 9.3, 7.9, 8.8,
#8.8, 12.7, 8.4, 7.8, 5.7, 10.5, 10.5, 9.6, 8.9,
#10.2, 10.3, 7.7, 10.6, 8.3, 8.8, 9.5, 8.8, 9.4])
#

### Problem 1.13.a

mean = 1/len(X) * np.sum(X)
s2 = (1/(len(X) - 1)) * (np.sum(X ** 2) - ((1/len(X)) * (np.sum(X) ** 2)))
stdd = np.sqrt(s2)

print(mean)
print(s2)
print(stdd)

### Problem 1.13.b

k1 = len(X) * 0.68
k2 = len(X) * 0.95
k3 = len(X)

ks = [k1, k2, k3]

for i in range(3):
  k = i+1

  front = mean - (k * stdd)
  end = mean + (k * stdd)

  Xfront = X > front
  Xend = X < end

  Xinterval = Xfront * Xend
  falls_in = np.sum(Xinterval)

  print(f'\nIn the interval [{front:.2f}:, {end:.2f}].')
  print(f'According to the Empirical Rule, I expect to find {ks[i]:.0f}')
  print(f'I found {falls_in:.0f} samples.\n')

print(f'The results do not follow the Empirical Rule because the Empirical Rule assumes my data obeys an approximately normal distribution.\n\
This is not the case because there is 1 extreme outliter sample in my data which causes my standard deviation to inflate.\n\
A more accurate analysis can be done by removing the outlier 35.1 value.')
```

```
##### Problem 1.15
plt.clf()
%clear
output.clear()

X = np.array([11.88, 6.27, 5.49, 4.81, 4.40, 3.78, 3.44, 3.11, 2.88, 2.68, 
7.99, 6.07, 5.26, 4.79, 4.05, 3.69, 3.36, 3.03, 2.74, 2.63, 
7.15, 5.98, 5.07, 4.55, 3.94, 3.62, 3.26, 2.99, 2.74, 2.62, 
7.13, 5.91, 4.94, 4.43, 3.93, 3.48, 3.20, 2.89, 2.69, 2.61])

# Removes the 11.88 outlier sample
#X = np.array([6.27, 5.49, 4.81, 4.40, 3.78, 3.44, 3.11, 2.88, 2.68, 
#7.99, 6.07, 5.26, 4.79, 4.05, 3.69, 3.36, 3.03, 2.74, 2.63, 
#7.15, 5.98, 5.07, 4.55, 3.94, 3.62, 3.26, 2.99, 2.74, 2.62, 
#7.13, 5.91, 4.94, 4.43, 3.93, 3.48, 3.20, 2.89, 2.69, 2.61])

### Problem 1.15.a

sum = 0
for sample in X:
  sum += sample

mean = sum / len(X)

sqr_devs = 0

for sample in X:
  sqr_devs += ((sample - mean) ** 2)

var = (1/(len(X) - 1)) * sqr_devs
stdd = np.sqrt(var)

print(mean)
print(stdd)

### Problem 1.15.b
k1 = len(X) * 0.68
k2 = len(X) * 0.95
k3 = len(X)

ks = [k1, k2, k3]

for i in range(3):
  k = i+1

  front = mean - (k * stdd)
  end = mean + (k * stdd)

  Xfront = X > front
  Xend = X < end

  Xinterval = Xfront * Xend
  falls_in = np.sum(Xinterval)

  print(f'\nIn the interval [{front:.2f}:, {end:.2f}].')
  print(f'According to the Empirical Rule, I expect to find {ks[i]:.0f}')
  print(f'I found {falls_in:.0f} samples.\n')

print(f'For lower values of k, I found more samples than expected. This is likely due to \
the outlier value 11.88 in the data. Removing that sample will cause the data to better fit the Empirical Rule.')
```
