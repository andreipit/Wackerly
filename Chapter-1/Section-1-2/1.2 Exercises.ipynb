```
##### Problem 1.2
plt.clf()
%clear
output.clear()

X = np.array([8.9, 12.4, 8.6, 11.3, 9.2, 8.8, 35.1, 6.2, 7.0,
7.1, 11.8, 10.7, 7.6, 9.1, 9.2, 8.2, 9.0, 8.7,
9.1, 10.9, 10.3, 9.6, 7.8, 11.5, 9.3, 7.9, 8.8,
8.8, 12.7, 8.4, 7.8, 5.7, 10.5, 10.5, 9.6, 8.9,
10.2, 10.3, 7.7, 10.6, 8.3, 8.8, 9.5, 8.8, 9.4])

###1.2.a
# Remove 35.1 from data
outliers = np.array([35.1])
no_outlierX = np.setdiff1d(X, outliers)

# Determine the endpoints of the histogram to not include 35.1
print(np.min(no_outlierX))
print(np.max(no_outlierX))

plt.hist(X, range=(5, 13), facecolor = '#2ab0ff', edgecolor='#169acf')

plt.title('Frequency Histogram of Avg Wind Speeds') 
plt.xlabel('Avg City Wind Speed (mph)') 
plt.ylabel('# Cities')

plt.savefig(f'{images_dir}/1_2_a.png')
plt.show()

### Problem 1.2.c
CHI_avg = 10.3

windier = X > CHI_avg

perc_windier = np.sum(windier) / len(X) * 100

print(f'{perc_windier:.2f}% of cities are windier than Chicago.')
```
