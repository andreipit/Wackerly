```python
##### Problem 1.24
plt.clf()
%clear
output.clear()

X = np.array([87, 109, 79, 80, 96, 95, 90, 92, 96, 98, 101, 91, 78, 112, 94, 98, 94, 107, 81, 96])

### Problem 1.24.a
min = np.min(X)
max = np.max(X)

rnge = max - min
stdd_approx = rnge/4

print(f'The Empirical Rules standard deviation approximation is {stdd_approx:.2f}.')

### Problem 1.24.b

plt.hist(X, facecolor = 'b', edgecolor = 'k')

plt.title(r'Histogram of Heart Rates During Aqua Running') 
plt.xlabel(r'Heart Rate') 
plt.ylabel('# People')

plt.savefig(f'{images_dir}/1_24_a.png')
plt.show()

### Problem 1.24.c
mean = np.mean(X)
stdd = np.std(X)

print(f'\nMean: {mean:.2f}')
print(f'Stdd: {stdd:.2f}\n')

### Problem 1.24.d

emp_rule = [0.68, 0.95, 1.0]

for i in np.arange(3):
  k = i + 1
  front = mean + (k * stdd)
  back = mean - (k * stdd)

  expected = len(X) * emp_rule[i]
  actual = np.sum((X < front) * (X > back))

  print(f'\n\nIn interval [{back:.2f}, {front:.2f}]')
  print(f'Expect: {expected:.2f} / {len(X)} samples found')
  print(f'{actual} / {len(X)} samples found')
```

```python
##### Problem 1.35
plt.clf()
%clear
output.clear()

X = np.array([172, 140, 123, 130, 115, 148, 108, 129, 137, 161, 123, 152, 133, 128, 142])

### Problem 1.35.a
stdd_approx = (np.max(X) - np.min(X))/4

print(f'I approximate standard deviation to be {stdd_approx:.2f}.')

### Problem 1.35.b
mean = np.mean(X)
stdd = np.std(X)

print(f'mean: {mean:.2f}')
print(f'stdd: {stdd:.2f}')

### Problem 1.35.c
print(f'By Tchebysheffs Innequality, to get 1 - 1/4 = 3/4 = 0.75 of the samples between some interval (a, b), \n\
we must choose k = 2 with the interval (mean - (k * stdd), mean + (k * stdd)). Thus our desired interval is \n\
({mean - (2 * stdd):.2f}, {mean + (2 * stdd):.2f}).\n\n')

a = mean - (2 * stdd)
b = mean + (2 * stdd)

print(f'Thus, a = {a:.2f} and b = {b:.2f}')

### Problem 1.35.d
lt = X < b
gt = X > a

between = np.sum(lt * gt)

print(f'I expect 0.75 of the points to be in the interval ({a:.2f}, {b:.2f}). \n\
{(between / len(X)):.2f} of the points are in the interval ({a:.2f}, {b:.2f}).\n\n')

print(f'Yes, our results are consistent with Tchebysheffs Innequality. In fact, the proportion of points \n\
in the interval was higher than 0.75. This is still consistent with Tchebysheffs Innequality because \n\
Tchebysheffs Innequality only gives us a lower bound on the proportion of the points in the interval.')

```
