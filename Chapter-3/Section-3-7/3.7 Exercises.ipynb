```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import hypergeom
from google.colab import output, drive

drive.mount('/content/gdrive/')
images_dir = '/content/gdrive/MyDrive/Colab_Notebooks/Wackerly/Chapter3/figures/Sec3_7/'
%matplotlib inline
```

```python
##### Problem 3.115
plt.clf()
%clear
output.clear()

freq = []

n = 3
r = 2
N = 6

Y = hypergeom(N, r, n)

for y in np.arange(0, 3):
  freq.append(Y.pmf(y))

freq = np.array(freq)

plt.bar(np.arange(3), freq, width=0.5, tick_label = np.arange(3),facecolor = '#2ab0ff', edgecolor='#169acf')

plt.title('Probability Histogram if Hypergeometric(3, 2, 6)') 
plt.xlabel('Y') 
plt.ylabel('Probability')

plt.savefig(f'{images_dir}/3_115.png')
plt.show()
```

```python
##### Problem 3.116
plt.clf()
%clear
output.clear()

n = 3
r = 2
N = 6

X = hypergeom.rvs(N, r, n, size = 100)

plt.hist(X, bins=3, facecolor = '#2ab0ff', edgecolor='#169acf')

plt.title('Frequency Histogram of 100 Samples from Hypergeometric(3, 2, 6)') 
plt.xlabel('Y') 
plt.ylabel('Frequency')

plt.savefig(f'{images_dir}/3_116.png')
plt.show()
```
