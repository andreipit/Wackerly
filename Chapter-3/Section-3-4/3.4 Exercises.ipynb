```python
##### Problem 3.46
plt.clf()
%clear
output.clear()

### p = .1
freq = []

for y in np.arange(0, 6):
  freq.append(binom.pmf(y, 5, .1))

freq = np.array(freq)

plt.bar(np.arange(6), freq, width=0.5, tick_label = np.arange(6),facecolor = '#2ab0ff', edgecolor='#169acf')

plt.title('Probability Histogram if B(5, .1)') 
plt.xlabel('Y') 
plt.ylabel('Probability')

plt.savefig(f'{images_dir}/3_46_p1.png')
plt.show()

### p = .5
plt.clf()
%clear

freq = []

for y in np.arange(0, 6):
  freq.append(binom.pmf(y, 5, .5))

freq = np.array(freq)

plt.bar(np.arange(6), freq, width=0.5, tick_label = np.arange(6),facecolor = '#2ab0ff', edgecolor='#169acf')

plt.title('Probability Histogram if B(5, .5)') 
plt.xlabel('Y') 
plt.ylabel('Probability')

plt.savefig(f'{images_dir}/3_46_p5.png')
plt.show()

plt.clf()
%clear

freq = []

for y in np.arange(0, 6):
  freq.append(binom.pmf(y, 5, .9))

freq = np.array(freq)

plt.bar(np.arange(6), freq, width=0.5, tick_label = np.arange(6),facecolor = '#2ab0ff', edgecolor='#169acf')

plt.title('Probability Histogram if B(5, .9)') 
plt.xlabel('Y') 
plt.ylabel('Probability')

plt.savefig(f'{images_dir}/3_46_p9.png')
plt.show()

print(f'Notice that the histogram for when p = .1 appears to be a reflection of the plot when p = .9 along the Y = 2.5 line (Y is on the horizontal axis).')
print(f'Also, it appears the histogram for when p = .5 is symmetric along the Y = 2.5 line.')
```
