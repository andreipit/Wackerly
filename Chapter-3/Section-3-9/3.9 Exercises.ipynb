```python
##### Problem 3.152
plt.clf()
%clear
output.clear()

prob = 0

for y in np.arange(11):
  prob += poisson.pmf(y, 6, 0)

for y in np.arange(2):
  prob -= poisson.pmf(y, 6, 0)

print(f'P(Y < 11) = {prob:.4f}')
```
