```python
##### Problem 3.121
plt.clf()
%clear
output.clear()

### Problem 3.121b
prob = 1

for y in np.arange(0, 4):
  prob -= poisson.pmf(y, 2)

print(f'P(Y>=4) = {prob:.4f}')

### Problem 3.121c
sum = 0

for y in np.arange(0, 4):
  sum += poisson.pmf(y, 2)

print(f'P(Y<4) = {sum:.4f}')

### Problem 3.121d
a = 1
b = 1

for y in np.arange(0, 4):
  a -= poisson.pmf(y, 2)

for y in np.arange(0, 2):
  b -= poisson.pmf(y, 2)

print(f'Probability Y >= 4 given Y >= 2 is: {a/b:.4f}')
```
